```python
#!venv/bin/python3
from CVRLoader import CVRLoader
from Contest import Contest
from elections.DiversityRunoff import *
from elections.InstantRunoffElection import InstantRunoffElection
from elections.HeadToHeadElection import HeadToHeadElection
from elections.PluralityElection import PluralityElection
from argparse import ArgumentParser
```

```python
primary_cvr = CVRLoader("2022Primary/CVR_Export_20220908084311", max_sessions=1000000)
# for each contest in the primary,   
#   print the name of the contest and the number of ballots.


def party_letter(party):
    if party.name == "Registered Democrat":
        return "D"
    elif party.name == "Registered Republican":
        return "R"
    elif party.name == "Registered Libertarian":
        return "L"
    elif party.name == "Registered AK Independence":
        return "A"
    elif party.name == "Nonpartisan":
        return "N"
    elif party.name == "Registered Veterans of AK":
        return "V"
    else:
        return "?"

# Create a list of contests with more than one rank
contests_without_multiple_ranks = [
    contest for contest in primary_cvr.elections.values()
    if primary_cvr.number_of_ranks_by_id[contest.contest_id] == 0
]

# Print the contests with more than one rank
for contest in contests_without_multiple_ranks:
    print(f"Contest Name: {contest.name}, "
          f"{primary_cvr.number_of_ranks_by_id[contest.contest_id]} "
          f"Number of Ballots: {len(contest.ballots)}"
          )
    
    # create a set of active_candidates for the contest by iterating over the ballots and flat-mapping the ordered_candidates
    active_candidates = set()
    for ballot in contest.ballots:
        for cs in ballot.ordered_candidates:
            active_candidates.add(cs.candidate)
        
    plurality = PluralityElection(contest.ballots, active_candidates)
    
    results = plurality.result()
    total_votes = sum(results.vote_totals.values())
    
    for cs in results.ordered_candidates:
        print("%40s" % cs.name, 
              "%5s" % party_letter(contest.parties_by_candidate[cs.id]),
              "%6d" % results.vote_totals[cs],
              "%6.2f" % (results.vote_totals[cs]/total_votes*100))
    
```
